```python
from statsmodels.tsa.stattools import adfuller
attribute_columns = ['att' + str(i) for i in range(1, 36)]
for attribute in attribute_columns:
    result = adfuller(df[attribute])

    test_statistic = result[0]
    p_value = result[1]

    print(f"\nDickey-Fuller Test Results for {attribute}:")
    print(f"Test Statistic: {test_statistic}")
    print(f"P-value: {p_value}")
if p_value <= 0.05:
        print("The data is likely stationary (reject the null hypothesis)")
else:
        print("The data is likely non-stationary (fail to reject the null hypothesis)")
```

```python
attribute_columns = df.columns[df.columns.str.startswith('att')]
df[attribute_columns] = (df[attribute_columns] - df[attribute_columns].min()) / (df[attribute_columns].max() - df[attribute_columns].min())
print(df)
```

```python
numeric_columns = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
column1 = df['att1']
column2 = df['att2']
t_statistic, p_value = stats.ttest_ind(column1, column2)
print("\nT-Test:")
print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The difference is statistically significant.")
else:
    print("The difference is not statistically significant.")
```
